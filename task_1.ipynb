```python
from os import path

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
population_data = pd.read_csv('/content/population.csv')  # Replace with the actual file path
#Example: Bar chart for population by country
population_data = population_data[["Country Name", "2020"]].dropna()  # Filter for 2020 data
population_data = population_data.rename(columns={"2020": "Population"})
population_data = population_data.nlargest(10, 'Population')  # Top 10 countries by population

plt.figure(figsize=(10, 6))
plt.bar(population_data["Country Name"], population_data["Population"], color='skyblue')
plt.title("Top 10 Countries by Population (2020)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
#print(population_data.head(5))
```
